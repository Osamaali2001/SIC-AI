```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("Employee.csv",sep = ',',encoding='UTF-8')
```

```python
data.head()
```

```python
import seaborn as sns
```

```python
# How is the Employee strength spread over office locations?
ax = sns.countplot(data=data,x="City")
for p in ax.patches:
   ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
```

```python
# does job location affects the decision of the employee whether leave or not?
ax = sns.countplot(x="City", hue="LeaveOrNot",data=data)
for p in ax.patches:
   ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.10, p.get_height()+0.01))
```

```python
# How does the company fare at Gender Diversity?
ax = sns.countplot(data=data,x="Gender")
for p in ax.patches:
   ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.30, p.get_height()+0.01))
```

```python
# does gender in job location affects the decision of the employee wther leave or not
g = sns.catplot(x="City", hue="LeaveOrNot",col="Gender",
                data=data, kind="count")
for ax in g.axes.ravel():
    for p in ax.patches:
     ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.10, p.get_height()+0.01))
```

```python
# How diverse are the Education levels of employees?
ax = sns.countplot(data=data,x="Education")
for p in ax.patches:
   ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.30, p.get_height()+0.01))
```

```python
# Does Employee's Education affect Payment Tiers?
ax = sns.countplot(x="PaymentTier",hue='Education', data=data)
for p in ax.patches:
   ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.10, p.get_height()+0.01))
```

```python
# Does Gender across Payment Tiers affect an Employee's decision whether he/she leaves the company or not?
g = sns.catplot(x="Gender", hue="PaymentTier",col="LeaveOrNot",
                data=data, kind="count")
for ax in g.axes.ravel():
    for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.10, p.get_height()+0.01))
```

```python
# how diverse are payment tiers for genders across cities?
g = sns.catplot(x="Gender", hue="PaymentTier",col="City",
                data=data, kind="count")
for ax in g.axes.ravel():
    for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.10, p.get_height()+0.01))
```

```python
# does gender has an impact on payment tier?
ax = sns.countplot(x="PaymentTier",hue='Gender', data=data)
for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.10, p.get_height()+0.01))
```

```python
# does payment tier affects employee decision on leaving or not
ax = sns.countplot(x="PaymentTier",hue='LeaveOrNot', data=data)
for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.10, p.get_height()+0.01))
```
